```python
# import libraries
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
import xgboost as xgb
from sklearn.metrics import roc_curve, precision_recall_curve
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns;sns.set()
```

## Importing the data
 - Importing the data is identical to the Cloud in this scenario
 - However, once the data was imported from the web from my local machine, I could 
   save the data to an object and use that saved data object for modeling. 
 - In the cloud environment this was quite different. First, was the data was imported from the web
   I had to save the data to a CSV file than upload the data to my S3 bucket.

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data', header = None)

# specify columns extracted from wbdc.names
data.columns = ["id","diagnosis","radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean",
                "compactness_mean","concavity_mean","concave points_mean","symmetry_mean","fractal_dimension_mean",
                "radius_se","texture_se","perimeter_se","area_se","smoothness_se","compactness_se","concavity_se",
                "concave points_se","symmetry_se","fractal_dimension_se","radius_worst","texture_worst",
                "perimeter_worst","area_worst","smoothness_worst","compactness_worst","concavity_worst",
                "concave points_worst","symmetry_worst","fractal_dimension_worst"] 

# save the data
data.to_csv("data.csv", sep=',', index=False)
```

## Crating a function that converts the target feature
 - M will equal zero
 - B will equal one
 - I used an identical function in the Cloud environment

```python
def target_encoder(dat):
    if dat == 'M':
        return 0
    else:
        return 1
data['diagnosis'] = data['diagnosis'].apply(lambda x:  target_encoder(x))
```

## Droping the ID column 

```python
df = data.drop(['id'],axis=1)
```

## Slitting the data into:
 - X is all features no target 
 - Y is target 

```python
# split dataset in features and target variable
X = df.drop(['diagnosis'], axis=1) # features
Y = data['diagnosis'] # target variable
```

## Splitting the data into test and train 
 - Splitting for Loistic Regression with L2 penalty 
 - Standarizing the test and train datasets 
 - I believe that in bost cases, building the model on my local computer and using AWS SageMaker the raw input data should be      preprocessed before making predictions.

```python
# split in train and test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=101)

# scale train dataset
scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)

# scale test dataset
X_test_scaled = scaler.transform(X_test)
```

## Tuning and Fitting the Logistic Regression model
 - On my local machine the logistic Regression can be trained and tuned in just two lines of code
 - For SageMaker it's quite a bit more involved. A list of some of the steps to building a model with SageMaker.
  - 1.) Create a SageMaker Instance and choose an instance type (ml.m5.large)
  - 2.) Inititate a SageMaker Session similar to: session = sagemaker.Session()
  - 3.) Create an IAM role for the SageMaker notebook instance. So SageMaker can have access to S3
  - 4.) Create an S3 Bucket for data upload & downloads and model storage 
  - 5.) I had to prepare the data for uploading to S3. Convert pandas dataframe to CSV file before uploading 
  - 6.) SageMaker algorithm image: after creating the SageMaker session and IAM role, I used the get_image_uri() funtion I think to get a Docker image. I wasn't 100% sure what all the Docker image did in all of this proccess. 
  - 7.) I build a dictionary of the hyper-parameters
  - 8.) Fit the model 
  - 9.) tested the model and got results
  
   

```python
# train logistic regression & cross validation
clf_L2 = LogisticRegressionCV(Cs=100, cv=10, penalty='l2', random_state=0, multi_class='multinomial', solver='lbfgs', n_jobs=-1)
clf_L2.fit(X_train_scaled, y_train)
```

## Getting prediction probs and metrics 
 - metrics for roc plot
 - metrics for precision Vs. recall
 - Evaluating the models using the ROC and precision Vs. recall was very similar in bothe cases 

```python
# Compute ROC curve and ROC area for each class
LR_y_score = clf_L2.predict_proba(X_test_scaled)[:,1]
LR_fpr, LR_tpr, LR_thresholds = roc_curve(y_test, LR_y_score)
```

 - If my feature importance method is correct, I think it's much easyer to do on my local machine using built in sklearn       functions and methods.

## My XGboost Model
 - Splitting the data into test and trian for xgboost model. 
 - Converting data with the DMatrix method 
 - The XGboost model with SageMaker was just as involved as the linear learner model 
 - Similar steps were taking to the onse that were listed above.
 - One difference between training the model on my local machine and SageMaker is I had to convert the data using a xgboost method cald DMatrx before I did any modeling of the data. I'm not 100% sure, but I could be posible that SageMaker might do this step before modeling autommatically before training.  

```python
# split in train and test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=101)



xg_trn_data = xgb.DMatrix(X_train, label=y_train)
xg_vld_data = xgb.DMatrix(X_test, label=y_test)
```

## Setting hyper-parameters
 - Setting the hyper-parameters seems pretty similar in both cases 
 - There dose seem to be some overfitting going on with the xgboost model in sklearn 
 - I did get to accually train and test my xgboost with SageMaker

```python
# XGBoost training
num_round = 50
xgb_param = {"objective": "binary:logistic" if xgb.__version__ > '0.82' else 'binary:logistic',
            'eta': 0.4, 'booster': 'gbtree', 'max_depth':3}

watchlist = [(xg_trn_data, "train"), (xg_vld_data, "valid")]

bst = xgb.train(xgb_param, xg_trn_data, num_round, watchlist)
bst
```

## A dictionary of all the features and I think the information gain score

```python
bst.get_score(importance_type='gain')
```

## Plotting the Feature importance for Xgboost 
 - plotting the feature importance on my local machine has a built in function that allows me to get a feature importance score
 - For SageMaker I don't believe there is a built in function to help plot the feature importance 
 - the variable radius_worst is highly significant compared to all other features

```python

feature_important = bst.get_score(importance_type='gain')
keys = list(feature_important.keys())
values = list(feature_important.values())

data = pd.DataFrame(data=values, index=keys, columns=["score"]).sort_values(by = "score", ascending=True)
data.plot(kind='barh')
```

## Plotting the Xgboost AUC

```python
ypred = bst.predict(xg_vld_data, ntree_limit=bst.best_ntree_limit)
xg_fpr, xg_tpr, xg_thresholds = roc_curve(y_test,ypred)
xg_roc_auc = auc(xg_fpr, xg_tpr)
print('xgboost AUC = ', xg_roc_auc)
```
