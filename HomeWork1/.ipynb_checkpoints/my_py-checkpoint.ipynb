# Quering S3 by unique key 

```python
# Selecting user_id, location, and added_key from each tweet where location equals 'Washington, D.C.' and 'Durham, NC'
query_list = []
time_list2 = []
for i in range(0,500):
    start = int(round(time.time() * 1000))
    r = s3.select_object_content(
        Bucket='dsba-6190-twitter',
        Key=f'twitter-json-files/tweet_unique_key_{i}.json',
        ExpressionType='SQL',
        Expression="SELECT s.user_id,s.location, s.added_key FROM s3object[*][*] s WHERE s.location IN ('Washington, D.C.','Durham, NC')",
        InputSerialization={'JSON': {"Type": 'DOCUMENT'}},
        OutputSerialization={'JSON': {}}   
    )
    end = int(round(time.time() * 1000))
    time_list2.append(end - start)
    query_list.append(r)
```

```python
# selecting user_id and added_key where added key is less than or equal to 50
for i in range(0,500):
    for event in query_list[i]['Payload']:
        if 'Records' in event:
            records = event['Records']['Payload'].decode('utf-8')
            print(records)
```

```python
# Listing the descriptive statistics of querying my S3 bucket 
s3_file_query_times1 = pd.DataFrame(time_list2,columns=['time_in_ms'])
s3_file_query_times1.describe()
```

```python
# Creating a distrabution plot of time each loop iteration when querying my s3 bucket 

s3_hist2 = sns.distplot(s3_file_query_times1['time_in_ms'],bins=25)
s3_hist2
```
